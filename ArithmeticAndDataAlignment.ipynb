```python
import pandas as pd
import numpy as np

s1 = pd.Series([0.7, 1.5, 2.2, 3.8], index=['a', 'b', 'c', 'd'])
s2 = pd.Series([4.9, 5.6, 6.0], index=['c', 'd', 'e'])

s1 + s2
# YOU CAN ADD THE TWO SERIES LIKE THIS IF THEY SHARE COMMON INDEX LABELS
# THE SAME CAN BE DONE WITH DATAFRAMES IF THEY SHARE THE SAME ROWS AND COLUMNS 
```

```python
# ADD TWO FULL DATAFRAMES TOGETHER

frame_one = pd.DataFrame(np.arange(12.).reshape((3,4)), columns=list('abcd'))
frame_two = pd.DataFrame(np.arange(20.).reshape((4,5)), columns=list('abcde'))

frame_one + frame_two
```

```python
# REPLACE THE NAN VALUES WITH ZERO

frame_one.add(frame_two, fill_value=0)
```

```python
frame_one.sub(frame_two, fill_value=0)
```

```python
frame_one.rdiv(frame_two)
```
